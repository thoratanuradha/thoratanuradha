# IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv("Training.csv.zip")
train
```

```python
test=pd.read_csv("Testing.csv")
test
```

# DROP UNWANTED COLUMN

```python
train=train.drop(["Unnamed: 133"],axis=1)
```

# SPLITTING DATESET

```python
P = train[["prognosis"]]
X = train.drop(["prognosis"],axis=1)
Y = test.drop(["prognosis"],axis=1)
```

### Separating the "Prognosis" column as the target variable (p) , and the rest of the columns as features (x) for training .Also,extracting features for the testing dataset (Y).

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,P,test_size=0.2,random_state=42)
```

### Splitting  the training and testing sets for model evaluation.

# MODEL - RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf= RandomForestClassifier(random_state=42)
model_rf = rf.fit(xtrain,ytrain)
tr_pred_rf = model_rf.predict(xtrain)
ts_pred_rf = model_rf.predict(xtest)

print("training accuracy is:",accuracy_score(ytrain,tr_pred_rf))
print("testing accuracy is:",accuracy_score(ytest,ts_pred_rf))
```

#### 1)Importing the RandomForestClassifier,setting up the model,and training it on the training dataset.
2) Making predictions on the training and testing datasets using  the trained model.
3) Printing the training and testing accuracy scores of the model.

# PREDICTIONS

```python
test.join(pd.DataFrame(model_rf.predict(Y),columns=["predicted"]))[["prognosis","predicted"]]
```
